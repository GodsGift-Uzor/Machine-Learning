```python
def KNN(x, y, x_q, K=3):
    target_labels = np.unique(y) # unique set of target labels
    target_labels_counts = np.zeros(len(target_labels)) # keeps counts of target labels
    d = np.sqrt(np.sum((x - x_q)**2, axis=1)) # use the Euclidean distance
    i = np.argsort(d) # sort distance vector in ascending order
    for k in np.arange(len(target_labels)):
        target_labels_counts[k] = np.sum(y[i[0:K]]==target_labels[k]) # count the number of each target lable in K Nearest Neighbourhood
    # apply the majority voting
    l = np.argmax(target_labels_counts)
    return target_labels[l]    
```

```python
N = y_testing.shape[0] # the number of examples in training dataset
print('{:>2} {:>16}'.format('K', 'Testing Error'))
for K in np.arange(1, 33, 2):
    y_prediction = np.zeros(N)
    for n in np.arange(N):
        x_q = x_testing[n]
        print(x_q)
        y_prediction[n] = KNN(x_training, y_training, x_q, K)
    classification_error = np.sum(y_testing != y_prediction) / N
    print('{:>2.0f} {:>16.4f}'.format(K, classification_error))
```

```python
class LogReg:
    def __init__(self, W):
        self.W = W
        self.delta = 0.00012 #change in error not to be exceeded
        self.epoch_max = 100000
        
    def parameters(self):
        return self.W
    
    def activation(self, s):
        #return 2.0 / (1 + np.exp(-2*s))-1 #between [-1,1]# sigmoid function as activation function
        return 1.0 / (1 + np.exp(-s))#[0,1]
    def activation_prime(self, s):
        return (self.activation(s) * (1.0 - self.activation(s))) # the derivative of activation function
    
    def feedforward(self, X):
        S = np.dot(X, self.W)
        U = self.activation(S)
        return (U)        
        
    def learn(self, X, Y, learning_rate):        
        C =  len(np.unique(Y))
        N = X.shape[0]
        #Yv = np.zeros((N, C))
        #Yv[np.arange(N), Y.astype(np.int)] = 1.0
        is_converged = False
        epoch = 1
        while not is_converged:
            U = self.feedforward(X)
            E = (U - Y)
            #print(E.shape)
            #print("asd ",np.dot(E, np.dot(U.T, (1-U))).shape)
            D = E * (U * (1-U))
            MSE = 1.0 / (2 * N) * np.sum(np.sum(E**2, axis=1))
            if (epoch%100) == 0:
                ACC = self.accuracy(X, np.asarray(pd.DataFrame(Y)[0]))
                print('epoch = {0}, MSE = {1:.3f}, ACC = {2:.3f}'.format(epoch, MSE, ACC))
            Wprev = self.W.copy()
            self.W = self.W - learning_rate * 1.0/N * np.dot(X.T, D)
            epoch += 1
            if (epoch >=self.epoch_max):
                is_converged = True 
        return self.W
        
    def classify(self, X):
        U = self.feedforward(X)
        L = []
        import pandas as pd
        for x in pd.DataFrame(U)[0].tolist():
            if(x >= 0.5):
                L.append(1);
            else:
                L.append(0);
        return pd.Series(L)
    
    def accuracy(self, X, Y):
        N = X.shape[0]
        L = self.classify(X)
        acc = np.sum(L == Y) / N
        return acc
    
    def confusion_matrix(self, X, Y):
        L = self.accuracy(X, Y)
        labels = np.unique(Y).astype(int)
        conf_mat = np.zeros((len(labels), len(labels)))
        for i in np.arange(len(labels)):
            for j in np.arange(len(labels)):
                conf_mat[i, j] = np.sum((L==labels[i]) and (Y==labels[j]))
```
